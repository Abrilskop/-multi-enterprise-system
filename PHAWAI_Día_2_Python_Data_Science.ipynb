## **Series**

Una serie es un objeto unidimensional similar a una matriz que contiene una secuencia de valores (de tipos similares a los tipos NumPy) del mismo tipo y una matriz asociada de etiquetas de datos, llamada índice. La Serie más simple se forma a partir de sólo una serie de datos usando `pd.Series()`:

````python
# Creando una serie a partir de una lista. En este caso no se proporcionan los índices (o etiquetas),
# entonces pandas crea los índices automáticamente

lst = ['a','b','c','d','e']
series = pd.Series(lst)
print(series)
````

Como se vió no hemos definido el índice y Pandas asignó como índice una secuencia `range` de números enteros desde el 0 hasta N-1.

Podemos especificar un índice usando un argumento opcional cuando construimos la serie.

````python
fruit_name_list = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
fruit_weight_list = [180, 120, 15, 650, 450]

my_fruit_series = pd.Series(data=fruit_weight_list,
                            index=fruit_name_list)
my_fruit_series
````

````python
# Creando una serie a partir de un diccionario. Las claves del diccionario se convierten en etiquetas y
# los valores de los elementos de la serie

dct = {'A' : 50, 'B' : 10, 'C' : 80, 'D' : 33}
series = pd.Series(dct)
print(series)
````

# Dataframes


Un Dataframe es un objeto de Pandas para organizar datos que actúa de manera muy similar a una hoja de cálculo.  Es bidimensional porque tiene filas y columnas.

Una Serie es una única columna de datos con un índice que identifica filas. Un marco de datos es una colección de series, que puede considerar como columnas, que comparten el mismo índice.

El índice compartido nos permitirá examinar, organizar y manipular muchos datos de manera muy eficiente.

![picture](https://drive.google.com/uc?export=view&id=1Vc-dO62jOh_35LMFpUPUf9MIzJXa7N5c)



Los Dataframes pueden ser creados de muchas maneras diferentes:
- __2-D NumPy array:__ Una matriz de datos, pudiendo pasar los índices de las filas y columnas
- __Dict of arrays, lists, or tuples:__ Cada secuencia se vuelve una columna. Las secuencias deben tener el mismo número de elementos.
- __Dict of Series:__ Cada serie se vuelve una columna. Los índices de cada serie son unidos para formar los índices de cada fila.
- __Dict of dicts:__ Cada diccionario se vuelve una columna. Las claves de los diccionarios se unen para formar los índices de las filas.
- __List of dicts or Series:__ Cada item se vuelve una línea en el DataFrame. La unidad de las claves (para dicionário) o índices (para Series) genera el  índice de las columnas.
- __List of lists or tuples:__  Similar a una matriz de numpy

##Concatenacion de Dataframes
**Concatenación de filas**. Las filas de los DataFrames se concatenan unas a continuación de las otras para formar el nuevo DataFrame. Para ello es necesario que los DataFrames que se combinen tengan el mismo índice de columnas.

pandas-concatenacion-filas.png



````python
import pandas as pd
df1 = pd.DataFrame({"Nombre":["Carmen", "Luis"], "Sexo":["Mujer", "Hombre"], "Edad":[22, 18]}).set_index("Nombre")
df2 = pd.DataFrame({"Nombre":["María", "Pedro"], "Sexo":["Mujer", "Hombre"], "Edad":[25, 30]}).set_index("Nombre")
df = pd.concat([df1, df2])
df
````

**Concatenación por columnas** Las columnas de los DataFrames se concatenan unas a continuación de las otras para formar el nuevo DataFrame. Para ello es necesario que los DataFrames que se combinen tengan el mismo índice de filas.
pandas-concatenacion-columnas.png

````python
import pandas as pd
df1 = pd.DataFrame({"Nombre":["Carmen", "Luis"], "Sexo":["Mujer", "Hombre"]}).set_index("Nombre")
df2 = pd.DataFrame({"Nombre":["Carmen", "Luis", "María"], "Edad":[22, 18, 25]}).set_index("Nombre")
df = pd.concat([df1, df2], axis = 1)
df
````

Además, también puede pasar una lista de columnas dentro de la notación de indexación `[]` para acceder a un subconjunto del marco de datos con solo las columnas dadas.

````python
# conjunto de columnas
casos_df = data_df[['departamento', 'edad', 'tipo_edad']]
casos_df
````

**IMPORTANTE** El nuevo data frame `casos_df` es simplemente una "view" del dataframe original` data_df`. Ambos apuntan a los mismos datos en la memoria de la computadora. Cambiar cualquier valor dentro de uno de ellos también cambiará los valores respectivos en el otro. Compartir datos entre dataframes  hace que la manipulación de datos en Pandas sea increíblemente rápida. No necesita preocuparse por la sobrecarga de copiar miles o millones de filas cada vez que desee crear un nuevo dataframe operando en uno existente.

A veces, es posible que necesite una copia completa del dataframe, en cuyo caso puede utilizar el método `copy`.

## Limpieza de Datos

 La naturaleza de los datos del mundo real incluye datos erroneos. Podría ser un error de entrada de datos, o el gobierno puede haber emitido una corrección para tener en cuenta los errores de conteo en el pasado.

````python
data_df.info()
````

````python
# vimos que tene,os que trabajar con localidad, diresa, localcod y unamed: 14 y 15
# podemos eliminarlas, reemplazar con 0 o con el promedio, media o moda.
# vamos a eliminar unnamed 14 y 15, localidad y localcod
````

Por ahora, supongamos que se trata de un error de entrada de datos. Podemos utilizar uno de los siguientes enfoques para tratar el valor faltante o defectuoso:
1. Reemplazar con "0".
2. Reemplazar con el promedio de toda la columna.
3. Reemplazar con el promedio de los valores de la fecha anterior y siguiente.
4. Descartar la fila por completo

El enfoque que elija requiere algo de contexto sobre los datos y el problema.


````python
# como no estan en mi documentacion, no hay infromacion de esos datos, entonces simplemente borramos toda la columna.
data_df.drop(['localidad', 'localcod', 'Unnamed: 14', 'Unnamed: 15'], axis = 1, inplace = True)
# inplace es para que elimin en el mismo dataframe no en la copia
````

También podemos formular consultas más complejas que involucran múltiples columnas.

````python
# Cantidad de casos por provincia en un departamento especifico
# aqui primero extraer quienes son los del departamento por ejemplo cusco, extraemos aquellos que son de cusco y de eso trabjamos
# usaremos el group by: agrupa y de ahi aplicamos cualquier funcion de aplicacion size la de suma, etc
# primero sacamos quienes pertenecen al departamento de cusco. agrupamos por provincia. y sacamos su tamaño
df_clean[df_clean.departamento == 'CUSCO'].groupby('provincia').size()
````

````python
# Cantidad de casos por provincia y por tipo de enfermedad
df_clean[df_clean.departamento == 'ANCASH'].groupby(['provincia', 'enfermedad']).size()
````

````python
#Crear funcion casos_enfermedad(departamento)
def casos_enfermedad(departamento):
    return df_clean[df_clean['departamento'] == departamento] \
             .groupby(['provincia', 'enfermedad']) \
             .size()
````
